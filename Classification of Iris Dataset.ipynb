```python
import tensorflow as tf
import numpy as np
import csv
import pandas as pd
```

```python
iris = pd.read_csv("iris.csv")
iris = iris.sample(frac=1).reset_index(drop=True)
iris.head(n =10)
```

```python
x_train = np.zeros((120,4),dtype = float)
y_train = np.zeros(120, dtype= int)
x_test = np.zeros((30,4),dtype = float)
y_test = np.zeros(30, dtype= int)
```

```python
for i in range(150):
    if i < 120:
        x_train[i][0] = iris.values[i][1]
        x_train[i][1] = iris.values[i][2]
        x_train[i][2] = iris.values[i][3]
        x_train[i][3] = iris.values[i][4]
        y_train[i] = check_species(iris.values[i][5])
    else:
        x_test[i-120][0] = iris.values[i][1]
        x_test[i-120][1] = iris.values[i][2]
        x_test[i-120][2] = iris.values[i][3]
        x_test[i-120][3] = iris.values[i][4]
        y_test[i-120] = check_species(iris.values[i][5])
```

```python
def check_species(y):
    if y == 'Iris-virginica':
        return 0
    elif y == 'Iris-versicolor':
        return 1
    else:
        return 2
```

```python
classes = ['Iris-virginica','Iris-versicolor','Iris-setosa']
x_train = tf.keras.utils.normalize(x_train, axis = 1) # normalizing from 0-255 to 0-1
x_test = tf.keras.utils.normalize(x_test, axis = 1)
```

```python
x_test
```

```python
model = tf.keras.models.Sequential() 
model.add(tf.keras.layers.Flatten()) ## input layer flattening the multi-dimension x_train into a 1-d array
model.add(tf.keras.layers.Dense(125, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(125, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(3, activation = tf.nn.softmax))# o/p layer where nodes equal to number of classes
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',metrics = ['accuracy']) 
                                        ## optimizer function, error measure and metrics
model.fit(x_train,y_train,epochs = 50)
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)
print(val_loss, val_acc)
```

```python
predictions = model.predict(x_test)
predictions[0]
```
